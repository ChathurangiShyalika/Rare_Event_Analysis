```python
import matplotlib.pyplot as plt4
import numpy as np
import pandas as pd
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

import tsaug
from tsaug import AddNoise, Convolve, Crop, Drift, Pool, Quantize, Resize, Reverse, TimeWarp
```

```python
df = pd.read_csv("bearing_dataset_combined_0.5%.csv") 
df.head(n=5)  # visualize the data.
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df
```

```python
#feature_drift_percentage 
# find the percentage change with the previous row
df2= df
df2=df2.pct_change()
df2=df2.rename(columns=
               {'DE': 'DE_pct_change','FE':'FE_pct_change'})
df2
```

```python
#feature_drift_percentage -PERIOD-5
# find the percentage change with the previous row
df22= df
df22=df22.pct_change(periods=5)
df22=df22.rename(columns=
               {'DE': 'DE_pct_change_5','FE':'FE_pct_change_5'})
df22
```

```python
#feature_drift_percentage -PERIOD-5
# find the percentage change with the previous row
df23= df
df23=df23.pct_change(periods=2)
df23=df23.rename(columns=
               {'DE': 'DE_pct_change_2','FE':'FE_pct_change_2'})
df23
```

```python
plot_acf(df['DE'], lags=700)
plot_pacf(df['DE'], lags=700)
```

```python
plot_acf(df['DE'], lags=10)
plot_pacf(df['DE'], lags=10)
```

```python
#lag features
df1=df
df1['lag_DE'] = df1['DE'].shift(1)
df1['lag_FE'] = df1['FE'].shift(1)


df3 = df1[['lag_DE', 'lag_FE']]
df3.head(10)
```

```python
#Rolling window

df1=df
df1['rolling_mean_DE'] = df1['DE'].rolling(window=7).mean()
df1['rolling_mean_FE'] = df1['FE'].rolling(window=7).mean()


df4 = df1[[ 'rolling_mean_DE', 'rolling_mean_FE']]
df4.head(10)
```

```python
df1 = pd.read_csv("bearing_dataset_combined_5%_augmented_stage1.csv") 
df1.head(n=5)  # visualize the data.
```

```python
#Convolve
df1['cnv_DE'] = tsaug.Convolve(window="flat", size=10).augment(df1.DE.to_numpy())
df1['cnv_FE'] = tsaug.Convolve(window="flat", size=10).augment(df1.FE.to_numpy())
df1
```

```python
#Pool
df1['pool_DE'] = tsaug.Pool(size=2).augment(df1.DE.to_numpy())
df1['pool_FE'] = tsaug.Pool(size=2).augment(df1.FE.to_numpy())
df1
```

```python
#Time warping
df1['tw_DE'] =tsaug.TimeWarp(n_speed_change=2, max_speed_ratio=5).augment(df1.DE.to_numpy())
df1['tw_FE'] =tsaug.TimeWarp(n_speed_change=2, max_speed_ratio=5).augment(df1.FE.to_numpy())
df1
```

```python
#Drifting
df1['drift_DE'] =tsaug.Drift(max_drift=0.3, n_drift_points=5).augment(df1.DE.to_numpy())
df1['drift_FE'] =tsaug.Drift(max_drift=0.3, n_drift_points=5).augment(df1.FE.to_numpy())
df1
```

```python
#STD-trend
from statsmodels.tsa.seasonal import seasonal_decompose

df1['trend_DE'] =seasonal_decompose(x=df1['DE'], model='additive', period=2).trend
df1['trend_FE'] =seasonal_decompose(x=df1['FE'], model='additive', period=2).trend
df1
```

```python
df1.to_csv('bearing_dataset_combined_5%_augmented_stage2.csv')
```
